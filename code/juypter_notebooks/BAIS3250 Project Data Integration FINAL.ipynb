```python
import pandas as pd
```

```python
scraped_salary = pd.read_csv('nfl_salary_cash_2020_2024.csv')
```

```python
scraped_salary
```

```python
team_data = pd.read_csv('nfl_team_data.csv')
team_data
```

```python
team_mapping = {
    "HOU": "Houston Texans",
    "NO": "New Orleans Saints",
    "DAL": "Dallas Cowboys",
    "LAC": "Los Angeles Chargers",
    "SF": "San Francisco 49ers",
    "PHI": "Philadelphia Eagles",
    "LV": "Las Vegas Raiders",
    "CIN": "Cincinnati Bengals",
    "MIA": "Miami Dolphins",
    "PIT": "Pittsburgh Steelers",
    "BAL": "Baltimore Ravens",
    "GB": "Green Bay Packers",
    "BUF": "Buffalo Bills",
    "TEN": "Tennessee Titans",
    "NYJ": "New York Jets",
    "SEA": "Seattle Seahawks",
    "KC": "Kansas City Chiefs",
    "LA": "Los Angeles Rams",
    "NE": "New England Patriots",
    "CHI": "Chicago Bears",
    "MIN": "Minnesota Vikings",
    "IND": "Indianapolis Colts",
    "ATL": "Atlanta Falcons",
    "CAR": "Carolina Panthers",
    "DET": "Detroit Lions",
    "JAX": "Jacksonville Jaguars",
    "NYG": "New York Giants",
    "WAS": "Washington Commanders",
    "ARI": "Arizona Cardinals",
    "CLE": "Cleveland Browns",
    "TB": "Tampa Bay Buccaneers",
    "LAR": "Los Angeles Rams",
    "DEN": "Denver Broncos"
}
```

```python
scraped_salary['Team Name'] = scraped_salary['Team'].replace(team_mapping)
scraped_salary
```

```python
name_correction = {
    "Washington Football Team": "Washington Commanders",
    "Washington Commanders": "Washington Commanders"
}
```

```python
team_data.rename(columns={'WT':'Winning Team Name','LT':'Losing Team','Season':'Year','WTS':'Winning Team Score','LTS':'Losing Team Score','DOW':'Day of Week'},inplace=True)
team_data
```

```python
team_data['Winning Team Name'] = team_data['Winning Team Name'].replace(name_correction)
team_data['Losing Team']=team_data['Losing Team'].replace(name_correction)
team_data
```

```python
# Merge salary info for winning teams
combined = pd.merge(
    team_data,
    scraped_salary,
    how='left',
    left_on=['Winning Team Name', 'Year'],
    right_on=['Team Name', 'Year']
)

# Rename winning team columns
combined = combined.rename(columns={
    'Rank': 'Winning Rank',
    'Team': 'Winning Team Code',
    'Record': 'Winning Record',
    'Signed Players': 'Winning Signed Players',
    'Avg Age': 'Winning Avg Age',
    'Active Cash': 'Winning Active Cash',
    'Dead Cash': 'Winning Dead Cash',
    'Total Cash': 'Winning Total Cash'
})

# Drop extra 'Team Name' from winning team join
combined = combined.drop(columns=['Team Name'])

# Merge salary info for losing teams
combined = pd.merge(
    combined,
    scraped_salary,
    how='left',
    left_on=['Losing Team', 'Year'],
    right_on=['Team Name', 'Year']
)

# Rename losing team columns
combined = combined.rename(columns={
    'Rank': 'Losing Rank',
    'Team': 'Losing Team Code',
    'Record': 'Losing Record',
    'Signed Players': 'Losing Signed Players',
    'Avg Age': 'Losing Avg Age',
    'Active Cash': 'Losing Active Cash',
    'Dead Cash': 'Losing Dead Cash',
    'Total Cash': 'Losing Total Cash'
})

# Drop extra 'Team Name' from losing team join
combined = combined.drop(columns=['Team Name'])
combined
```

```python
combined.to_csv("combined_nfl_data.csv")
```
